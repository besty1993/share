```python
import os
import sys
import time
import gc
import glob
import random

import numpy as np
import matplotlib.pyplot as plt
from skimage import io
from PIL import Image

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import torchvision
from torchvision import transforms
from torch.cuda.amp import autocast
from torchsummary import summary

import resnet
from efficientnet_pytorch import EfficientNet
from opt.Lookahead import Lookahead
from opt.Ralamb import Ralamb
```

```python
EPOCHS = 15
TRAIN_RATIO = 0.8
BATCH_SIZE = 256
SEED = 0

def seed_everything(seed):
    random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)

seed_everything(SEED)

train_path = '../data/train'
test_path = '../data/test'
train_list = os.listdir(train_path)
random.shuffle(train_list)
val_list = train_list[int(len(train_list)*TRAIN_RATIO):]
train_list = train_list[:int(len(train_list)*TRAIN_RATIO)]
test_list = os.listdir(test_path)
```

```python
transform=torchvision.transforms.Compose([
    torchvision.transforms.Resize((256,256)),
    torchvision.transforms.RandomHorizontalFlip(p=0.5),
    torchvision.transforms.RandomRotation(180),
    torchvision.transforms.ToTensor()
])

class DogCatDS_Train(Dataset) :
    def __init__(self, data_paths, transform=None) :
        self.data_paths = data_paths
        self.transform = transform
        
    def __len__(self) :
        return len(self.data_paths)
    
    def __getitem__(self, idx) :
        if torch.is_tensor(idx) :
            idx = idx.tolist()
        
        img_name = self.data_paths[idx]
        label = img_name.split('/')[-1].split('\\')[-1][:3]
        if label == 'cat' :
            label = 0
        else :
            label = 1
        image = Image.open(img_name)
        
        if self.transform :
            image = self.transform(image)
            
        sample = {'image' : image, 'label': label}
        return sample
    
class DogCatDS_Test(Dataset) :
    def __init__(self, data_paths) :
        self.data_paths = data_paths
        self.transform = torchvision.transforms.Compose([
            torchvision.transforms.Resize((256,256)),
            torchvision.transforms.ToTensor()
        ])
        
    def __len__(self) :
        return len(self.data_paths)
    
    def __getitem__(self, idx) :
        if torch.is_tensor(idx) :
            idx = idx.tolist()
        
        img_name = self.data_paths[idx]
        image = Image.open(img_name)
        image = self.transform(image)
        label = 0
        return {'image' : image, 'label' : label}
    

trainset = DogCatDS_Train(train_list, transform=transform)
valset = DogCatDS_Train(val_list, transform=transform)
testset = DogCatDS_Test(test_list)

trainloader = DataLoader(trainset, batch_size=BATCH_SIZE, shuffle=True)
valloader = DataLoader(valset, batch_size=BATCH_SIZE, shuffle=False)
testloader = DataLoader(testset, batch_size=BATCH_SIZE, shuffle=False)
```

```python
fig = plt.figure(figsize=(20,5))
for i, sample in enumerate(trainloader) :
    print(i, sample['image'].shape, len(sample['label']))
    
    if len(sample['image'].shape) == 4 :
        img = np.moveaxis(sample['image'].numpy()[0], 0, -1)
    
        ax = fig.add_subplot(1,4,i+1)
        ax.imshow(img)
        ax.axis('off')
        ax.set_title(sample['label'][0])
    
    if i >=3 :
        plt.show()
        break
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu") # PyTorch v0.4.0
print(device)
```

```python
#### EfficientNet-B0 ####################
model = EfficientNet.from_pretrained('efficientnet-b0')
model._fc = torch.nn.Linear(in_features=model._fc.in_features, out_features=2, bias=True)
model.to(device)
summary(model, (3,256,256))

#### SE-ResNeXt101-32x4d ####################
# load_state = torch.load('../weights/nvidia_se-resnext101-32x4d_200821.pth.tar')
# new_load_state = load_state.copy()
# for key in load_state.keys() :
#     n_key = key[7:]
#     new_load_state[n_key] = new_load_state.pop(key)
    
# model = resnet.build_resnet('se-resnext101-32x4d', 'classic', 1000)
# model.load_state_dict(new_load_state)
# model.fc = torch.nn.Linear(in_features=model.fc.in_features, out_features=2, bias=True)
# model.to(device)
# summary(model, (3,256,256))
# del load_state, new_load_state
```

```python
criterion = nn.CrossEntropyLoss()
# optimizer = optim.Adam(model.parameters())
ralamb = Ralamb(model.parameters(), weight_decay=0.01)
optimizer = Lookahead(ralamb)
scheduler = optim.lr_scheduler.CosineAnnealingWarmRestarts(optimizer, 10, 3, eta_min=1e-6)
```

```python
%%time

model.train()
for epoch in range(EPOCHS) :
    print(f'Epoch : {epoch+1}')
    stime = sstime = time.time()
    running_loss = val_running_loss = 0.0
    
    #### Training ####
    for i, data in enumerate(trainloader) :
        inputs, labels = data['image'].to(device), data['label'].to(device)

        optimizer.zero_grad()
        
        with autocast() :
            outputs = model(inputs)
            loss = criterion(outputs, labels)
        
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        print(f'{i+1}/{len(trainloader)} steps ({time.time()-sstime:.2f}sec/step)', end='\r')
        sstime = time.time()
    
    #### Validation ####
    for j, data in enumerate(valloader) :
        inputs, labels = data['image'].to(device), data['label'].to(device)
        
        with autocast() :
            outputs = model(inputs)
            loss = criterion(outputs, labels)
        
        val_running_loss += loss.item()        
    
    ## Update Learning Rate ##
    scheduler.step()
    
    print('loss : {:.4f}, val_loss : {:.4f}, taken time : {:.2f}sec'.format(
        running_loss/(i+1), val_running_loss/(1+1), time.time()-stime)
    )
```

```python
%%time

model.eval()

fig = plt.figure(figsize=(20,5))
with torch.no_grad() :
    for i, data in enumerate(testloader) :
        inputs, labels = data['image'].to(device), data['label'].to(device)
        
        with autocast():
            output = model(inputs)

        ax = fig.add_subplot(1,4,i+1)
        img = np.moveaxis(inputs[0].cpu().numpy(), 0, -1)
        ax.imshow(img)
        ax.axis('off')
        ax.set_title(output[0].cpu().tolist())

        if i >= 3 :
            plt.show()
            break
```

```python
class_correct = list(0.0 for i in range(2))
class_total = list(0.0 for i in range(2))
with torch.no_grad() :
    for data in valloader :
        inputs, labels = data['image'].to(device), data['label'].to(device)
        with autocast() :
            outputs = model(inputs)
        _, predicts = torch.max(outputs,1)

        c = (predicts==labels).squeeze()
        for i in range(BATCH_SIZE) :
            try :
                label = labels[i]
                class_correct[label] += c[i].item()
                class_total[label] += 1
            except Exception as e :
                pass

classes = ['cat', 'dog']
for i in range(2) :
    print('Accuracy of {} : {}'.format(
        classes[i], 100*class_correct[i]/class_total[i]
    ))
```
